### **Q2: What is the range of BU GPA for student athletes?**


```python
import pandas as pd
import matplotlib.pyplot as plt

# file_path
file_path = r'C:\Users\chenk\OneDrive\桌面\BU Study\DS701\Q2\cleaned_data.csv'

# Read the data file
data = pd.read_csv(file_path)

# Filter out 'Cum GPA' columns, keeping only 'GPI' columns
gpi_columns = [col for col in data.columns if 'GPI' in col]
gpi_data = data[gpi_columns]

# Calculate the average GPA for each row, ignoring NaN values, and store it back in the original DataFrame
data['average_gpa'] = gpi_data.mean(axis=1, skipna=True)

# Calculate the statistics
gpa_min = data['average_gpa'].min()
gpa_max = data['average_gpa'].max()
gpa_range = gpa_max - gpa_min
q1 = data['average_gpa'].quantile(0.25)
median = data['average_gpa'].median()
q3 = data['average_gpa'].quantile(0.75)

# Print the statistics
print("Minimum Average GPA:", gpa_min)
print("Maximum Average GPA:", gpa_max)
print("GPA Range:", [gpa_min,gpa_max])
print("First Quartile (Q1):", q1)
print("Median (Q2):", median)
print("Third Quartile (Q3):", q3)

# create and show a box plot for average GPA
plt.figure(figsize=(8, 6))
plt.boxplot(data['average_gpa'].dropna(), vert=False)
plt.title('Range of BU GPA for student athlete')
plt.xlabel('Average GPA')
plt.grid(axis='x', linestyle='--')
plt.show()
```
